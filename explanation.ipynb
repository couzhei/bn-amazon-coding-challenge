```python
from amazon_challenge.utility import *
```

We'll create the vacant 10x10 (n = 10) space, using `generate_graph()` (the default value is `n_nodes=10`):

```python
my_graph = generate_graph(n_nodes=10)
```

We can show the said graph with the defined function `plot_results` giving this graph as an input:

```python
plot_results(my_graph)
```

Now setting our starting point to (0,0) and destination to (9,9) will result the following:

```python
start = (0,0)
destination = (9,9)

plot_results(my_graph, start=start, destination=destination)
```

Showing all the available nodes (units of plane) and all the paths by setting `show_nodes` and `show_paths` to `True`:

```python
plot_results(my_graph, start=start, destination=destination, show_nodes=True, show_paths=True)
```

Now putting some obstacles at (9, 7), (8, 7), (6, 7), and (6, 8), the space and its possible paths changes as a consequence. We can easily create obstacles and show them via `plot_results`:

```python
obstacles = {(9, 7), (8, 7), (6, 7), (6, 8)}
plot_results(generate_graph(), obstacles=obstacles, start=start, destination=destination, show_paths=True)
```

At any time, if you want to see the shortest path from the starting point to the destination, you just need to set `show_results` to `True`:

```python
plot_results(my_graph, obstacles=obstacles, start=start, destination=destination,
            show_results=True)
```

Generating 20 more obstacles will results the following:

```python
list(map(lambda x:tuple(map(int, tuple(x.rstrip(")").lstrip("(").split(",")))), "(1,2) (2,4)".split()))
```

```python
import random
random.seed(42)
obstacles = {(9, 7), (8, 7), (6, 7), (6, 8)}

for i in range(20):
    while True:
        candidate = randint(0, 9), randint(0, 9)
        if candidate in obstacles | {start} | {destination}:
            continue
        else:
            obstacles.add(candidate)
            break

plot_results(my_graph, obstacles=obstacles, start=start, destination=destination,
            show_results=True)
```

This time `show_results=True` gives us a warning that the destination is unreachable.

In order to see which minimum number of obstacles can be removed so the delivery point becomes reachable once again we use `remove_obstacles()` function as follows:

```python
removed_obstacles, dist, path = remove_obstacles(my_graph, obstacles, start, destination)
```

Now we can give the key word argument `to_be_removed` set to be `removed_obstacles`:

```python
plot_results(my_graph, obstacles=obstacles, start=start, destination=destination,
            show_results=True, to_be_removed = removed_obstacles)
```

What if we give it another point to be removed (instead of the optimal solution)? It still works this way:

```python
the_other_point = {(1,0),}
```

```python
plot_results(my_graph, obstacles=obstacles, start=start, destination=destination,
            show_results=True, to_be_removed = the_other_point)
```
